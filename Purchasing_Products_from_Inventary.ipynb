## Purchasing Products from Inventary

```python
bill={}
total_amount=0

print("*************************************Welcome in Shop*****************************************")

# Displaying Menu
print("\n*********************************************************************************************")
for i in records.keys():
    print(i, records[i])
    
print("\n**********************************************************************************************\n")


while True:
# Taking Input from User to Enter what he/she wants to purchase
    ui_prod  = (str(input("\nEnter the product_id you want to purchase: ")))
    ui_quant = (int(input("Enter the quantity you want: ")))
    ui=str(input("\nYou Want to purchase more items Y/N:"))
    
    if ui_prod not in records.keys():                                                                       #Checking if product Exists in Inventary or not
        print("Product not in Inventary")
        
    elif records[ui_prod]['quantity']==0:                                                                  #Checking if user entered product quantity is zero
        print("Sorry,Currently We dont have stock for this product")
        if ui=='N' or ui=='n':
            break  
        
    else:
        if (ui_quant > records[ui_prod]['quantity']):                                                                            #Checking if user entered product quantity greater than available quantity
            print("Sorry, we have only",records[ui_prod]['quantity'],"quantity left in our stock." )       
            print("You want",records[ui_prod]['quantity'],"quantity")
            i=input("Press Y/N:")   
            
            if i=='Y' or i=='y':                                                                                                                     #if user want to purchase quantity we're having
                bill[len(bill)+1]={"Product":records[ui_prod]['name'],"Price":records[ui_prod]['price'],"Quantity":records[ui_prod]['quantity']}     #Adding into bill 
                records[ui_prod]['quantity'] = 0                                                                                                     #updating inventary
                sales[len(sales)+ 1]={'product' : ui_prod, 'quantity' : ui_quant, 'amount': records[ui_prod]['price'] * ui_quant,'Time':time.ctime()}       
                print("\n*************************************************************************")
                print("                                    BILL                                 ")
                print("-------------------------------------------------------------------------")
                print("Product","\t Price","\t Quantity","\t Total")
                for i in bill:
                    print(bill[i]['Product'],"\t\t",bill[i]['Price'],"\t",bill[i]["Quantity"],"\t\t",bill[i]["Quantity"]*bill[i]["Price"])
                print("-------------------------------------------------------------------------")
                total_amount=bill[i]["Quantity"]*bill[i]["Price"]+total_amount
                print("Total Amount :\t\t\t\t",total_amount)
                print("\n*************************************************************************")
                if ui=='N' or ui=='n':
                    break
                    
            else:                                                                                                                #if user dont want that much quantity
                print("Sorry,Currently we don't have that much quantity in our Stock")
                if ui=='N' or ui=='n':
                    break
            
                
        else:                                                                                              
            bill[len(bill)+1]={"Product":records[ui_prod]['name'],"Price":records[ui_prod]['price'],"Quantity":ui_quant}         #Adding into bill 
            records[ui_prod]['quantity'] = records[ui_prod]['quantity'] - ui_quant                                               #updating inventary
            sales[len(sales)+ 1]={'product' : ui_prod, 'quantity' : ui_quant, 'amount': records[ui_prod]['price'] * ui_quant,'Time':time.ctime()}       
            print("\n*************************************************************************")
            print("                                    BILL                                 ")
            print("-------------------------------------------------------------------------")
            print("Product","\t Price","\t Quantity","\t Total")
            for i in bill:
                print(bill[i]['Product'],"\t\t",bill[i]['Price'],"\t",bill[i]["Quantity"],"\t\t",bill[i]["Quantity"]*bill[i]["Price"])
            print("-------------------------------------------------------------------------")
            total_amount=bill[i]["Quantity"]*bill[i]["Price"]+total_amount
            print("Total Amount :\t\t\t\t",total_amount)
            print("-------------------------------------------------------------------------")
            print("Thank You! Visit Again.")
            print("\n*************************************************************************")
            if ui=='N' or ui=='n':
                    break
       

            
            
rec = json.dumps(records)                          # Updating Sales File

fd = open("record.json",'w')
fd.write(rec)
fd.close()


sale = json.dumps(sales)                         # Updating Inventory File

fd = open("sales.json",'w')
fd.write(sale)
fd.close()
            
```
