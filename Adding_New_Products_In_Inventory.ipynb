## Reading Data From JSON file

```python
import json
```

```python
fd = open("record.json",'r')
r = fd.read()
fd.close()

record = json.loads(r)
```

```python
record
```

## Add New Items in Inventory

```python
n=int(input("How many products you want to add in Inventary: "))
for i in range(n):
    prod_id = str(input("\nEnter product id: "))
    if prod_id in record.keys():
        qn = int(input("Enter quantity: "))
        record[prod_id]['quantity'] = record[prod_id]['quantity']+qn

    else:
        name = str(input("Enter product name: "))
        company = str(input("Enter Product Company: "))
        pr_size = str(input("Enter product size: "))
        pr_price = int(input("Enter product price: "))
        qn = int(input("Enter quantity: "))
        record[prod_id] = {'name': name, 'company':company, 'size': pr_size, 'price': pr_price, 'quantity':qn}

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
